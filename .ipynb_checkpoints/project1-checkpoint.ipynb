```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv('placement.csv')
```

```python
data
```

```python
#Preprocessing

data = data.iloc[:,1 :]
```

```python
data
```

```python
# EDA 
plt.scatter(data['cgpa'], data['iq'], c= data['placement'] )
```

```python
# Extract input and output 

x = data.iloc[:,:2]  # Independent variable 
y = data.iloc[:, 2]   # Dependent variable 
```

```python
x
```

```python
y
```

```python
# train the test split 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)  #0.1 means 10% of data means 10 rows can goes 
```

```python
x_test
```

```python
x_train
```

```python
y_test
```

```python
y_train
```

```python
# scale the value 
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
# Model training 

clf = LogisticRegression()
```

```python
clf.fit(x_train, y_train)
```

```python
# Evaluate the model 

y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
accuracy = accuracy_score(y_pred, y_test)
```

```python
f"The accuracy of our model is {accuracy*100}%."
```

```python
# visualizing the model pattern using decision boundary
```

```python
plot_decision_regions(x_train, y_train.values, clf=clf, legend=2)
```
